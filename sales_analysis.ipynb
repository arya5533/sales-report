```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv("sales.csv")

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Add Total column (Price √ó Quantity)
df['Total'] = df['Price'] * df['Quantity']

# Show DataFrame
print("üìä Full Sales Data:")
print(df)

# Total revenue per product
revenue_by_product = df.groupby('Product')['Total'].sum()
print("\nüí∞ Total Revenue by Product:")
print(revenue_by_product)

# Best-selling product
best_product = revenue_by_product.idxmax()
best_revenue = revenue_by_product.max()
print(f"\nüèÜ Best-selling product: {best_product} (${best_revenue})")

# Monthly sales trend
df['Month'] = df['Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Total'].sum()

# Plot monthly sales
monthly_sales.plot(kind='line', marker='o', title='Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
